```python
import pandas as pd

df = pd.read_csv('data/cars.csv')
df.head()
```

A coluna **id** e **link** não representa relevancia na nossa analise. A coluna **condition** a maioria dos dados são nulos por isso resolvi dropa-la. Todos os tipos dos carros são sedan também, por esse motivo podemos remover a coluna **type**.

```python
df.drop(['id', 'link', 'condition', 'type'], axis = 1, inplace=True)
```

```python
df.head()
```

```python
df.info()
```

Variaveis quantitativas:

miles, city_mileage, highway_mileage, horsepower, torque, fuel_capacity, price, front_headroom, front_legroom, rear_headroom, rear_legroom

Variaveis qualitativas:

engine_capacity_litre, num_cylinder, num_seat, num_owners, doors, wheel_drive, engine_type, speed_levels, service_records

Temos um dado nulo apenas no nosso dataframe referente ao id 7, na coluna speed_levels, e temos apenas dois carros com o tipo de motor elétrico id 7 e 18. Resolvi remover esses dados para não interferir para compararmos apenas carros de combustível a gás. logo, iremos deletar as linhas de id 7 e 18 e dropar a coluna engine_type, visto que todos os carros possuem motor com combustível a gás. 

```python
df.drop([7,18], inplace=True)
```

```python
df.drop('engine_type', inplace=True, axis=1)
```

```python
df = df.reset_index(drop=True)
```

Todos os carros também possuem 4 portas e tem 5 assentos, logo não faz sentido termos essa coluna no dataset.

```python
df.drop('doors', inplace=True, axis=1)
```

```python
df.drop('num_seat', inplace=True, axis=1)
```

Temos também apenas um carro com 6 cilindros na nossa amostra, seria conveniente remover essa linha e retirarmos também essa coluna para simplificar nossa análise.

```python
df.drop([18], inplace=True)
```

```python
df.drop('num_cylinder', inplace=True, axis=1)
```

```python
df.head()
```

Veremos se as variaveis do DataFrame possuem relação com a variável preço por meio de uma EDA:

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

Primeiramente vamos avaliar as variaveis quantitativas (engine_capacity_litre, num_owners, wheel_drive, speed_levels, service_records) em relação ao preço

```python
fig, ax = plt.subplots(figsize=(8,6))
sns.boxplot(x='num_owners', y= 'price',data = df)
plt.title('Relação entre o preço e o numero de proprietários')
plt.show()
```

Neste gráfico, vemos que quanto maior o numero de proprietários, mais o carro tende a ser mais barato.

```python
fig, ax = plt.subplots(figsize=(8,6))
sns.boxplot(x='wheel_drive', y= 'price',data = df)
plt.title('Relação entre o preço e Tração do carro')
plt.show()
```

Os carros tracionados nas 4 rodas, tendem a ser mais caros que os carros com tração em 2 rodas.

As demais colunas quantitativas, não trouxeram insights em relação ao preço.

```python
fig, ax = plt.subplots(4,2,figsize=(12,12))
plt.subplot(4,2,1)
sns.scatterplot(data=df, y='price',x='miles')
plt.subplot(4,2,2)
sns.scatterplot(data=df, y='price',x='horsepower')
plt.subplot(4,2,3)
sns.scatterplot(data=df, y='price',x='torque')
plt.subplot(4,2,4)
sns.scatterplot(data=df, y='price',x='fuel_capacity')
plt.subplot(4,2,5)
sns.scatterplot(data=df, y='price',x='front_headroom')
plt.subplot(4,2,6)
sns.scatterplot(data=df, y='price',x='front_legroom')
plt.subplot(4,2,7)
sns.scatterplot(data=df, y='price',x='rear_headroom')
plt.subplot(4,2,8)
sns.scatterplot(data=df, y='price',x='rear_legroom')
```

1. Quanto mais milhas, menor o preço do carro;
2. Quanto mais cavalos, maior o preço do carro;
3. Quanto mais torque, maior o preço do carro;
4. Quanto ao espaço interno de altura e espaço para pernas não parece ter muita relação com o preço do veiculo

--------

Veremos se esses levantamentos fazem sentido na matriz de correlação. 

```python
matriz_correlacao = df.corr()
fig, ax = plt.subplots(figsize=(12,12))
sns.heatmap(data = matriz_correlacao, annot=True)
```

Pelo que pode ser visto parece que quantop mais moderno o carro, maior é seu preõ, também podemos dizer que quanto mais milhas rodadas o preço do carro tende a abaixar, o torque e a potencia do motor (horsepower) parece não influenciar tanto quanto pensanmos de inicio. De fato também podemos dizer que quanto mais proprietários aquele carro teve, o preço irá ser menor e curiosamente carros que possuem maior espaço para as pernas no banco traseiro também tendem a ter o preço mais elevado.
