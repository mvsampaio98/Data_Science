```python
import pandas as pd                      
import matplotlib.pyplot as plt          
import seaborn as sns                    
import numpy as np                        

from sklearn.model_selection import train_test_split 
from sklearn.metrics import accuracy_score 

from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree  
from sklearn import tree

from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import export_graphviz
```

```python
dados = pd.read_csv("data/card_transdata.csv", sep=",")
```

```python
dados.head()
```

```python
dados.isnull().sum() 
```

```python
# Separando dados:

x = dados.drop(columns=['fraud'])
y = dados['fraud'] #O que eu quero prever. (Target)
```

```python
# Separando bases de treino e teste:
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, stratify=y, random_state=7) 
```

```python
# Criando modelo de arvore:

dt = DecisionTreeClassifier(random_state=7, criterion='entropy', max_depth = 2)
dt.fit(x_train, y_train)
```

```python
y_predito = dt.predict(x_test) 
```

```python
tree.plot_tree(dt)
```

```python
class_names = ['Fraude', 'Não Fraude']
label_names = ['distance_from_home', 'distance_from_last_transaction', 'ratio_to_median_purcharse_price', 'repeat_retailer', 'used_chip', 'used_pin_number', 'online_order']

fig, axes = plt.subplots(nrows = 1, ncols = 1, figsize = (15,15), dpi = 300)

tree.plot_tree(dt, feature_names=label_names, class_names=class_names, filled = True)
```

----------

```python
rf = RandomForestClassifier(n_estimators = 5, max_depth = 2, random_state=7)

rf.fit(x_train, y_train)
```

```python
estimator = rf.estimators_
```

```python
y_predito_random_forest = rf.predict(x_test)
```

```python
print(accuracy_score(y_test,y_predito_random_forest))
```

```python
class_names = ['Fraude', 'Não Fraude']
label_names = ['distance_from_home', 'distance_from_last_transaction', 'ratio_to_median_purcharse_price', 'repeat_retailer', 'used_chip', 'used_pin_number', 'online_order']
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 1, figsize = (15,15), dpi = 300)

tree.plot_tree(rf.estimators_[0], feature_names=label_names, class_names=class_names, filled = True)
```

```python
fig, axes = plt.subplots(nrows = 1,ncols = 5,figsize = (10,2), dpi=900)
for index in range(0, 5):
    tree.plot_tree(rf.estimators_[index],
                   feature_names = label_names, 
                   class_names=class_names,
                   filled = True,
                   ax = axes[index]);

    axes[index].set_title('Estimator: ' + str(index), fontsize = 11)
```

```python
print (rf.score(x_train, y_train)) 
print(rf.score(x_test, y_test))
```
