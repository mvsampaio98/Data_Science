### Importando a base de dados no Pandas:

```python
import pandas as pd
```

Utilizamos o *encoding* correto para a leitura dos nossos dados, o parâmetro *skiprows* para pular as primeiras linhas desnecessárias e o *skipfooter* para pular as últimas linhas. Os parâmetros *thousands* e *decimal* é utilizado para trocar as casas decimais de vírgula para ponto por exemplo, ficando igual a convenção americana de dados numéricos. 

```python
dados = pd.read_csv('dados/sih_cnv_qiuf214900192_141_121_80.csv', encoding='ISO-8859-1', skiprows=3,sep=';',skipfooter=12, thousands=',', decimal='.')
```

### Tendo uma primeira visualização da produção de medicamentos por Unidade de Federação:

```python
dados.plot(x='Unidade da Federação', y='2023/Ago')
```

### Esse gráfico não é adequado para nosso objetivo, o mais adequado é um gráfico de barras conforme abaixo:

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
axis = dados.plot(kind='bar',x='Unidade da Federação', y='2023/Ago', figsize=(12,7))
axis.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))

plt.title('Produção Hospitalar por Unidade de Federação em Agosto de 2023')
plt.show()
```

### Deixando as legendas com uma visualização melhor:

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
axis = dados.plot(kind='bar',x='Unidade da Federação', y='2023/Ago', figsize=(12,7))
axis.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))

plt.title('Produção Hospitalar por Unidade de Federação em Agosto de 2023')
plt.xticks(rotation=60,ha='right')
plt.show()
```

Importantes pontos a se considerar na hora de construir gráficos:
    
1. Mostrar os dados;
2. Levar o espectador a pensar sobre a substância em vez de metodologia, design gráfico, tecnologia de produção gráfica ou qualquer outra coisa;

4. Apresentar muitos números em um espaço pequeno;
5. Tornar grandes conjuntos de dados coerentes;
6. Encorajar o olho a comparar diferentes pedaços de dados;
7. Revelar os dados em vários níveis de detalhe, desde uma visão geral ampla até a estrutura fina;
