### Importando a base de dados no Pandas:

```python
import pandas as pd
```

Utilizamos o *encoding* correto para a leitura dos nossos dados, o parâmetro *skiprows* para pular as primeiras linhas desnecessárias e o *skipfooter* para pular as últimas linhas.

```python
dados = pd.read_csv('dados/sih_cnv_qiuf214900192_141_121_80.csv', encoding='ISO-8859-1', skiprows=3,sep=';',skipfooter=12)
```

### Transformando para o padrão americano numérico:

```python
dados = dados.apply(lambda x: x.str.replace(',','.'))
```

### Mudando o índice do DataFrame:

```python
dados_indexados = dados.set_index('Unidade da Federação')
```

### Removendo uma inconsistência de dados:

```python
dados_indexados = dados_indexados[~(dados_indexados == '-')]
```

### Transformando as colunas em float:

```python
dados_indexados = dados_indexados.astype('float')
```

### Criando uma função para criar uma coluna com as regiões de cada Estado:

```python
def regioes(x):
    if x == '35 São Paulo' or x == '31 Minas Gerais' or x == '33 Rio de Janeiro' or x == '32 Espírito Santo':
        return 'Sudeste'
    elif x == '41 Paraná' or x == '42 Santa Catarina' or x == '43 Rio Grande do Sul':
        return 'Sul'
    elif x == '52 Goiás' or x == '51 Mato Grosso' or x == '50 Mato Grosso do Sul' or x == '53 Distrito Federal':
        return 'Centro-Oeste'
    elif x == '16 Amapá' or x == '14 Roraima' or x == '12 Acre' or x == '17 Tocantins' or x == '11 Rondônia' or x == '13 Amazonas' or x == '15 Pará':
        return 'Norte'
    else:
        return 'Nordeste'
```

### Adicionando uma coluna com a região de cada Estado:

```python
dados_indexados['Unidade da Federação'] = dados_indexados.index
dados_indexados['Região'] = dados_indexados['Unidade da Federação'].apply(regioes)
dados_indexados = dados_indexados.drop('Unidade da Federação', axis=1)
dados_indexados
```

### Refinando o gráfico obtido na aula anterior:

```python
dados_sem_regiao = dados_indexados.drop('Região', axis = 1)
dados_sem_regiao = dados_sem_regiao / 1000000
```

```python
axis = dados_sem_regiao.head().T.plot(figsize=(12,6))
plt.title('Produção Hospitalar entre Jan/2008 até Ago/2023')
plt.xlabel('Anos/Meses')
plt.ylabel('Gasto aprovado em milhões de reais')
plt.legend(ncols=1, fontsize='medium')
plt.show()
```

```python
dados_sem_regiao[dados_sem_regiao.columns[0:181]].head(3).T.plot(figsize=(12,6),grid = True,color=['green','blue','red'])
plt.ylim(0,600)
plt.xticks(range(0, 181, 12), dados_sem_regiao.columns[0:181:12], rotation=45)
plt.title('Gasto com Produção Hospitalar entre Jan/2008 até Ago/2023')
plt.xlabel('Anos')
plt.ylabel('Gasto aprovado em milhões de reais')
plt.show()
```

Pontos importantes na construção de um gráfico:

1. Titulos e Legendas;
2. Rótulos de eixos;
3. Escala;
4. Cores (Destacar informações e tornar o gráfico mais atraente);
5. Marcadores (Para indicar pontos específicos no gráfico;
6. Grade (Facilitar a leitura do gráfico);
7. Estilo de linha.

### Pegando o mês mais atual e fazendo um plot, dando destaque para meu estado:

```python
colors = []
gastos_mais_recente = dados_sem_regiao[dados_sem_regiao.columns[-1]]
gastos_mais_recente = gastos_mais_recente.sort_values(ascending=False)
for i in gastos_mais_recente:
    if i > gastos_mais_recente['29 Bahia']:
        colors.append('lightgreen')
    elif i < gastos_mais_recente['29 Bahia']:
        colors.append('coral')
    else:
        colors.append('gold')
gastos_mais_recente.plot(kind='bar', figsize=(12,6),grid = True, color=colors)
plt.xticks(rotation=60, ha='right')
plt.title('Gasto com Produção Hospitalar em Ago/2023 em milhões de reais')
plt.ylabel('Gasto aprovado em milhões de reais')
plt.xlabel('')
plt.axhline(y=gastos_mais_recente['29 Bahia'], color='gold', linestyle='--')
plt.show()
```

### Escolha dois estados e plote os gastos deles de acordo com sua população:

```python
dois_estados = gastos_mais_recente[['35 São Paulo', '29 Bahia']]
dois_estados['35 São Paulo'] = dois_estados['35 São Paulo'] / 44.420
dois_estados['29 Bahia'] = dois_estados['29 Bahia'] / 14.136
dois_estados.plot(kind='bar',figsize=(12,6),grid = True,color=['lightgreen','red'])
plt.title('Gasto com Produção Hospitalar em Ago/2023 por população')
plt.ylabel('Gasto aprovado em milhões de reais por população')
plt.xlabel('')
plt.xticks(rotation=0, ha='center')
plt.show()
```
