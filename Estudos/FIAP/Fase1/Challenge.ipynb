### Agora precisamos obter os valores entre 2007-2022:

```python
def tratar_dados_imet():
    dados_imet_rs_prec_lista = []
    dados_imet_rs_umd_lista = []
    for i in range(0,16):
        dados_imet_rs = pd.read_csv(f'dados/INMET_S_RS_A801_PORTO ALEGRE_{2007+i}.csv', encoding='ISO-8859-1', sep=';',skiprows=8)
        dados_imet_rs['PRECIPITAÇÃO TOTAL, HORÁRIO (mm)'] = dados_imet_rs['PRECIPITAÇÃO TOTAL, HORÁRIO (mm)'].str.replace(',','.')
        dados_imet_rs['PRECIPITAÇÃO TOTAL, HORÁRIO (mm)'] = dados_imet_rs['PRECIPITAÇÃO TOTAL, HORÁRIO (mm)'].astype('float')
        dados_imet_rs = dados_imet_rs[dados_imet_rs['PRECIPITAÇÃO TOTAL, HORÁRIO (mm)'] >= 0]
        dados_imet_rs_prec_lista.append(dados_imet_rs['PRECIPITAÇÃO TOTAL, HORÁRIO (mm)'].sum())
        dados_imet_rs_umd_lista.append(dados_imet_rs['UMIDADE RELATIVA DO AR, HORARIA (%)'].mean())
    return dados_imet_rs_prec_lista, dados_imet_rs_umd_lista
```

### Obtendo todas as somas de precipitação entre 2007 e 2022 em Porto Alegre:

```python
dados_imet_rs_listas = tratar_dados_imet()
dados_imet_rs_listas
```

### Criando um DataFrame:

```python
dic_imet = {'Ano': np.arange(2007,2023), 'Precipitacao_Total': dados_imet_rs_listas[0], 'Umidade_Media_%': dados_imet_rs_listas[1]}
df_imet_tratado = pd.DataFrame(data = dic_imet)
df_imet_tratado
```

### Visualizando a precipitação em um gráfico de barras:

```python
import seaborn as sns

plt.figure(figsize=(12,6))
plt.grid(True)
sns.barplot(data=df_imet_tratado, x='Ano', y='Precipitacao_Total')
plt.axhline(df_imet_tratado['Precipitacao_Total'].mean(), color='r')
plt.title('Precipitação Total (mm) por ano')
plt.xlabel(' ')
plt.ylabel(' ')
plt.show()
```

Consegui notar uma precitação auge no período de  2014 e 2015, que pode ter influenciado diretamente na produção de vinhos e derivados no ano de 2016, e por isso pode ter acontecido um queda brusca de produção no ano de 2016

### Visualizando a umidade em um gráfico de barras:

```python
import seaborn as sns

plt.figure(figsize=(12,6))
plt.grid(True)
plt.title('Porcentagem de Umidade média por ano')
plt.xlabel(' ')
plt.ylabel(' ')
sns.barplot(data=df_imet_tratado, x='Ano', y='Umidade_Media_%')
plt.axhline(df_imet_tratado['Umidade_Media_%'].mean(), color='r')
plt.show()
```

Nesse gráfico esperava ver uma maior umidade nos anos de 2015 e 2016, a umidade interfere diretamente na proliferação de pragas e no nascimento dos vinhedos. Entretanto, parece se manter constante e não consegui extrair mais insights sobre esse ponto.
