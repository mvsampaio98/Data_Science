### Importando base de dados e visualizando as primeiras linhas:

```python
import pandas as pd

dados = pd.read_csv('dados/producao.csv', sep=';', encoding='ISO-8859-1')
dados.head()
```

### Checando o tamanho do DataFrame:

```python
print(f'O tamanho do Dataset é de {dados.shape[0]} linhas e {dados.shape[1]} colunas')
```

### Checando os tipos de dados do DataFrame:

```python
dados.info()
```

### Dropar a coluna ID:

```python
dados_gerais = dados.drop('id', axis=1)
dados_gerais.head()
```

### Checar a coluna produto:

```python
dados_gerais['produto']
```

### Filtrando o DataFrame apenas com os dados gerais e desprezando as subcategorias nesse momento:

```python
dados_gerais = dados_gerais.query('(produto == "VINHO DE MESA") | (produto == "VINHO FINO DE MESA (VINÃFERA)") | (produto == "SUCO") | (produto == "DERIVADOS")')
dados_gerais = dados_gerais.reset_index(drop=True)
dados_gerais
```

### Definindo produto como Index:

```python
dados_gerais = dados_gerais.set_index('produto')
dados_gerais
```

### Renomeando o Indice:

```python
dados_gerais.index = ['VINHO DE MESA', 'VINHO FINO DE MESA', 'SUCO', 'DERIVADOS']
dados_gerais
```

### Primeiras visualizações:

```python
import matplotlib.pyplot as plt
import numpy as np

dados_gerais_milhares = dados_gerais / 1000000
dados_gerais_milhares.T.loc['2007':].plot(figsize=(12,6))
plt.xticks(np.arange(0, 16, step=1))
plt.title('Produção em Milhões de Litros de produtos feitos com uva')
plt.ylabel('Milhões de Litros')
plt.grid()
plt.show()
```

**Por que houve uma queda de produção brusca no ano de 2016 ?**

Segundo essa [fonte](https://www.embrapa.br/busca-de-noticias/-/noticia/9103859/artigo-safra-da-uva-2016----o-que-esta-acontecendo) ocorreu o seguinte:

**Eventos de granizo, geadas e o excesso de chuvas, determinado pelo fenômeno climático El Niño, têm causado uma perda significativa na produtividade dos vinhedos em todas as áreas produtoras de uvas do RS. Tudo começou com um inverno de temperaturas amenas(30 a 40% de horas de frio, comparado com o normal), fazendo brotar menos ramos nas videiras.**
