## 5. Tabalhando com Banco de Dados

```python
import pandas as pd
import sqlalchemy
```

```python
sqlalchemy.__version__
```

### Criando um banco de dados local

```python
from sqlalchemy import create_engine, MetaData, Table, inspect
```

```python
engine = create_engine('sqlite:///:memory:')
```

```python
engine
```

### Escrevendo em um banco de dados

```python
url = 'https://raw.githubusercontent.com/alura-cursos/Pandas/main/clientes_banco.csv'
```

```python
dados = pd.read_csv(url)
```

```python
dados.head()
```

```python
# Salvando a tabela com o nome clientes dentro da nossa engine que foi criada
dados.to_sql('Clientes', engine, index = False)
```

```python
# Essa função permite acessarmos as tabelas da engine
inspector = inspect(engine)
```

```python
# Ver quais os nomes das tabelas presentes no nosso banco de dados
print(inspector.get_table_names())
```

### Lendo uma consulta SQL

```python
query = 'SELECT * FROM Clientes WHERE Categoria_de_renda = "Empregado"'
```

```python
# Para fazer a consulta precisamos usar a função read_sql e passar como 
# parâmetro a consulta que desejamos (query) e nosso banco de dados (engine)
empregados = pd.read_sql(query, engine)
```

```python
empregados.head()
```

```python
# Para salvar a tabela utilizar to_sql, passando o parâmetro 'con' igual a
# nossa engine que foi criada
empregados.to_sql('empregados', con=engine, index=False)
```

```python
# read_sql_table para ler uma tabela inteira, passando a engine que está nossa
# tabela.
pd.read_sql_table('empregados', engine)
```

```python
# Podemos também visualizar só algumas colunas da tabela
pd.read_sql_table('empregados', engine, columns=['ID_Cliente', 'Grau_escolaridade', 'Rendimento_anual'])
```

### Atualizando um banco de dados

```python
query = 'SELECT * FROM Clientes'
```

```python
pd.read_sql(query, engine)
```

```python
# Queremos deletar a primeira linha do nosso banco de dados.
# Usamos o connect para conectar com o banco de dados e o execute faz com que
# execute a query que criamos
query = 'DELETE FROM Clientes WHERE ID_Cliente =5008804'
with engine.connect() as conn:
    conn.execute(query)
```

```python
pd.read_sql_table('Clientes', engine)
```

```python
# Queremos agora atualizar o cadastro de um cliente que mudou seu grau
# de escolaridade de ensino médio para ensino superior
query = 'UPDATE Clientes SET Grau_escolaridade="Ensino Superior" WHERE ID_Cliente=5008808'
with engine.connect() as conn:
    conn.execute(query)
```

```python
pd.read_sql_table('Clientes', engine)
```

### Hora da Prática

Você é responsável por criar um banco de dados local de clientes para uma instituição financeira. Temos o arquivo CSV com os dados de clientes.

Sua missão é:

1.Criar o banco de dados local com a biblioteca SQLAlchemy.

2.Escrever os dados do arquivo CSV neste banco de dados local.

3.Realizar três atualizações no banco de dados:

Atualizar o registro do cliente de ID 6840104 que teve o rendimento anual alterado para 300000.

Excluir o registro do cliente de ID 5008809, pois essa pessoa não possui mais conta na instituição financeira.

Criar um novo registro de cliente seguindo as especificações abaixo:
ID_Cliente: 6850985
Idade: 33
Grau_escolaridade: Doutorado
Estado_civil: Solteiro
Tamanho_familia: 1
Categoria_de_renda: Empregado
Ocupacao: TI
Anos_empregado: 2
Rendimento_anual: 290000
Tem_carro: 0
Moradia: Casa/apartamento próprio

```python
url = 'https://raw.githubusercontent.com/alura-cursos/Pandas/main/clientes_banco.csv'
```

```python
dados
```

```python
engine2 = create_engine('sqlite:///:memory:')
```
